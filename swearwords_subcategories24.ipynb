```python
import pandas as pd
import numpy as np
```

```python

#train_df = pd.read_csv("dataset/hindi_alltasks_metadata.csv")
train_df=pd.read_csv("dataset/hindi_iter2_subcategories.csv")
train_df= train_df.dropna(how='all')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Raw data (your fine categories)
# -----------------------------
sizes = [
        count_Swear_Explicit_Bollywood_Personalities,
        count_Swear_Explicit_Cricket_Personalities,
        count_Swear_Explicit_Historical_Figures,
        count_Swear_Explicit_Media_Personalities,
        count_Swear_Explicit_Other_Personalities,
        count_Swear_Explicit_Politician,
        
        count_Swear_Explicit_Caste_Groups,
        count_Swear_Explicit_Religious_Groups,
        count_Swear_Explicit_Political_Groups,
        count_Swear_Explicit_Other_Groups,
        
        count_Swear_Implicit_Hindus,
        count_Swear_Implicit_Muslims,
        count_Swear_Implicit_Others,
        count_Swear_Implicit_Personalised_Political_Slurs,
        count_Swear_Implicit_Political_Group_Slurs
]
labels = [
    "SwearWords Explicit Bollywood Personalities",
    "SwearWords Explicit Cricket Personalities",
    "SwearWords Explicit Historical Figures",
    "SwearWords Explicit Media Personalities",
    "SwearWords Explicit Other Personalities",
    "SwearWords Explicit Politician",
        
        
    "SwearWords Explicit Caste Groups",
    "SwearWords Explicit Religious Groups",
    "SwearWords Explicit Political Groups",
    "SwearWords Explicit Other Groups",
        
        
    "SwearWords Implicit Hindus",
    "SwearWords Implicit Muslims",
    "SwearWords Implicit Others",
    "SwearWords Implicit Personalised Political Slurs",
    "SwearWords Implicit Political Group Slurs"
]

# -----------------------------
# Parent categories
# -----------------------------
group_of = {
    "Explicit Target Groups": [
        "SwearWords Explicit Caste Groups",
        "SwearWords Explicit Other Groups",
        "SwearWords Explicit Religious Groups",
        "SwearWords Explicit Political Groups",
    ],
    "Implicit Target": [
        "SwearWords Implicit Personalised Political Slurs",
        "SwearWords Implicit Hindus",
        "SwearWords Implicit Political Group Slurs",
        "SwearWords Implicit Others",
        "SwearWords Implicit Muslims",
    ],
    "Explicit Target Personalities": [
        "SwearWords Explicit Cricket Personalities",
        "SwearWords Explicit Media Personalities",
        "SwearWords Explicit Bollywood Personalities",
        "SwearWords Explicit Historical Figures",
        "SwearWords Explicit Other Personalities",
        "SwearWords Explicit Politician",
    ],
}

size_dict = dict(zip(labels, sizes))

inner_labels = list(group_of.keys())
inner_sizes = [sum(size_dict[x] for x in group_of[g]) for g in inner_labels]

inner_colors = ["#f4a582", "#92c5de", "#b8e186"]

# -----------------------------
# Outer ring inherits parent color
# -----------------------------
outer_labels = []
outer_sizes = []
outer_colors = []

for g, parent_color in zip(inner_labels, inner_colors):
    for child in group_of[g]:
        outer_labels.append(child)
        outer_sizes.append(size_dict[child])
        outer_colors.append(parent_color)

fig, ax = plt.subplots(figsize=(16, 16))

# -----------------------------
# INNER RING (now with labels)
# -----------------------------
wedges_inner, texts_inner = ax.pie(
    inner_sizes,
    labels=inner_labels,      # <- label parent categories
    radius=0.4,
    startangle=90,
    colors=inner_colors,
    labeldistance=0.3,        # place labels inside the ring
    textprops={'fontsize': 8,'fontfamily':'Arial','fontweight':'bold'},
    wedgeprops=dict(width=0.3, edgecolor="white"),
)

# -----------------------------
# OUTER RING (with percentages)
# -----------------------------
wedges_outer, texts_outer, autotexts_outer = ax.pie(
    outer_sizes,
    labels=outer_labels,
    radius=0.7,
    startangle=90,
    labeldistance=1.05,       # labels outside
    autopct='%1.1f%%',        # show percentages
    pctdistance=0.8,          # percentage text inside the ring
    textprops={'fontsize': 9,'fontfamily':'Arial','fontweight':'bold'},
    colors=outer_colors,
    wedgeprops=dict(width=0.3, edgecolor="white"),
)

ax.set(aspect="equal")
#ax.set_title("Proportion of the tweets with the appropriate subcategories",pad=0,fontsize=20,fontweight='bold')

# -----------------------------
# LEGEND FOR INNER CIRCLE (optional, keeps color mapping clear)
# -----------------------------
legend_labels = [
    "Explicit Target Groups",
    "Implicit Target",
    "Explicit Target Personalities"
]

ax.legend(
    wedges_inner,
    legend_labels,
    title="Main Categories",
    loc="upper right",
    #bbox_to_anchor=(1.25, 1.0)
    bbox_to_anchor=(0.2, 0.3),
    title_fontproperties={'size': 16, 'weight': 'bold'}  # legend title
)

#plt.subplots_adjust(top=0.50)
plt.tight_layout()
plt.show()
```
