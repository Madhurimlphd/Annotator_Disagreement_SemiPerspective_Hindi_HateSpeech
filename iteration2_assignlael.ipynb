
In this notebook, <br/>


1] The tweets were assigned new annotations performed in "relabelling iteration2" by matching tweet IDs from the original dataset with the Iteration2 (hate/non-hate) ids and storing the results in a new column, New_Label_iter2. <br/>

2] The [New_label] column stores relabeled annotations from iteration 1. For tweets that were not relabeled in this iteration, the values from [Original_label] are copied into [New_label]. <br/>

3] The [New_label_iter2] column stores relabeled annotations from iteration 2. For tweets that were not relabeled in this iteration, the values from the [New_label] column are copied into [New_label_iter2]. <br/>

4] The [Mismatch] column was assigned a value of 1 if either (A) the original annotations stored in [Original Label] differed from the iteration-1 relabeling annotations stored in [New_Label], or (B) the original annotations stored in [Original Label] differed from the iteration-2 relabeling annotations stored in [New_Label_iter2].


```python
import pandas as pd
import numpy as np
```

```python

train_df = pd.read_csv("dataset/Hindi_Alltasks.csv")
train_df= train_df.dropna(how='all')
```

```python
train_df.head(20)
```

```python
mismatch_count = (train_df["Mismatch"] == 1).sum()
print("Count of relabelled tweets before iteration2 relabelling is",mismatch_count)
```

IDs of relabelled tweets that should be labeled as hate.

```python
#hate_ids=[ 1441, 2419, 2420, 2506,2760, 2768, 2784, 3223,3474,3751,4378,4424 ] (2420 already relabelled HOF in first iteration)
hate_ids=[ 1441, 2419, 2506,2760, 2768, 2784, 3223,3474,3751,4378,4424 ]
```

```python
#### create the label column if it doesn't exist
####train_df["New_label_iter2"] = None

##### compute the doubled hate_ids
double_hate_ids = [(i - 2) for i in hate_ids]

#### assign "HOF" where train_df index matches 2 * hate_ids
train_df.loc[train_df.index.isin(double_hate_ids), "New_label_iter2"] = "HOF"
```

```python
print("The size of Iteration 2 tweets that have been relabelled as HOF.",len(train_df[train_df["New_label_iter2"]=="HOF"]))
```

IDs of relabelled tweets that should be labeled as NOT

```python
non_hate_ids=[1120,1125,1192,1415,1567,1828, 1848, 1875,1895,1916,1941, 1950, 1988, 1993, 1996, 2299,2862,2940,2944,2963,2984,3061,3087, 3102,3198,3375, 3485,3534,3573,3694,3750,3804,3922, 3935, 3976, 4216,4279, 4283,4563 ]
#1565 already labelled NOT in iteration1
```

```python


##### compute the doubled hate_ids
double_nonhate_ids = [(i - 2) for i in non_hate_ids]

##### assign "HOF" where train_df index matches 2 * hate_ids
train_df.loc[train_df.index.isin(double_nonhate_ids), "New_label_iter2"] = "NOT"
```

```python
print("The size of Iteration 2 tweets that have been relabelled as NOT",len(train_df[train_df["New_label_iter2"]=="NOT"]))
```

Filling the empty value in New_label_iter2

```python
train_df["New_label_iter2"] = train_df["New_label_iter2"].fillna(train_df["New_label"])
#The New_label_iter2 column stores relabeled annotations from iteration 2. For tweets that were not relabeled in this iteration, the values from the New_label column are copied into New_label_iter2.
```

```python
train_df[train_df['Mismatch'] == 1].head(10)
#Top 10 rows of relabelled iteration1 (mismatch=1, at this stage) to check whether New_label is assigned to empty values in New_label_iter2.
```

```python
train_df.loc[train_df.index.isin(double_hate_ids)]
#Display the relabeled iteration-2 hate values (New_label_iter2) to check whether they are correct and did not altered by the existing New_label values.”
#new_label_iter2 (relabelled iteration2) should be different from New_label(relabelled iteration1)
#mismatch column values have not yet been updates, so all rows have value=0
```

Finding the disagreement tweets

```python
train_df['Mismatch'] = (
    (train_df['Original_label'] != train_df['New_label_iter2']) |
    (train_df['Original_label'] != train_df['New_label'])
).astype(int)
```

```python
mismatch_count = (train_df["Mismatch"] == 1).sum()
print("Count of relabelled tweets after iteration1 and iteration2",mismatch_count)
```

Saving the dataset to the file

```python
train_df.to_csv("dataset/Hindi_alltasks_iteration2label.csv",index=False)
```
